```python
base_color = sns.color_palette()[0]
plt.figure(figsize = (16, 12))
sns.countplot(data = twitter_archive_master, 
              y = twitter_archive_master.breed,
              order = twitter_archive_master.breed.value_counts()[:30].index, 
              color = base_color,
             );
for i in range(30):
    count = twitter_archive_master.breed.value_counts()[i]
    pct_string = '{}'.format(count)
    plt.text(count + 1, i, pct_string, va='center')
    
plt.title('top 30 breed')
```

```python
top30breed = twitter_archive_master.breed.value_counts()[:30].index
twitter_top30 = twitter_archive_master[twitter_archive_master.breed.isin(top30breed)]

twitter_top30_avg = twitter_top30\
.groupby('breed')\
.mean()
```

```python
base_color = sns.color_palette()[0]
plt.figure(figsize = (14, 8))
sns.barplot(data = twitter_top30_avg, 
              x = 'favorite_count', 
              y = twitter_top30_avg.index,
              order = twitter_top30_avg.sort_values('favorite_count', ascending = False).index,
              color = base_color,
             );
for i in range(30):
    count = twitter_top30_avg.sort_values('favorite_count', ascending = False)['favorite_count'][i]
    pct_string = '{:0.2f}'.format(count)
    plt.text(count + 1, i, pct_string, va='center')

plt.title('avg favorite_count for top 30 breed')
```

```python
base_color = sns.color_palette()[0]
plt.figure(figsize = (14, 8))
sns.barplot(data = twitter_top30_avg, 
              x = 'retweet_count', 
              y = twitter_top30_avg.index,
              order = twitter_top30_avg.sort_values('retweet_count', ascending = False).index,
              color = base_color,
             );
for i in range(30):
    count = twitter_top30_avg.sort_values('retweet_count', ascending = False)['retweet_count'][i]
    pct_string = '{:0.2f}'.format(count)
    plt.text(count + 1, i, pct_string, va='center')
    
plt.title('avg retweet_count for top 30 breed')
```

```python
twitter_top30_avg['rating'] = twitter_top30_avg['rating_numerator']/twitter_top30_avg['rating_denominator']
base_color = sns.color_palette()[0]
plt.figure(figsize = (14, 8))
sns.barplot(data = twitter_top30_avg, 
              x = 'rating', 
              y = twitter_top30_avg.index,
              order = twitter_top30_avg.sort_values('rating', ascending = False).index,
              color = base_color,
             );
for i in range(30):
    count = twitter_top30_avg.sort_values('rating', ascending = False)['rating'][i]
    pct_string = '{:0.2f}'.format(count)
    plt.text(count +0.02, i, pct_string, va='center')

plt.title('avg rating for top 30 breed')
```

#### Question 2: Twitter Analyics -- Twitter account development

From the number of tweets posted on the @WeRateDog account, these is a peak around end of 2015, then the posts decrease next several months, and then gets relatively stational post number after April 2016. <br>

For number of favorites, we could tell the favorites increase from Jan 2016 to June 2017, and get a shape drop after June 2017.<br>

For number of retweet, it shows the retweet gets its peak around the end of 2015, and then get a small peak around May 2016 and Jan 2017.

```python
twitter_archive_master['date'] = pd\
.to_datetime(twitter_archive_master['timestamp'])\
.dt.to_period('D')


twitter_archive_master['year_month'] = pd\
.to_datetime(twitter_archive_master['timestamp'])\
.dt.to_period('M')

trend_by_date = twitter_archive_master.groupby('date').agg({'tweet_id': 'count',
                                           'favorite_count': 'sum',
                                           'retweet_count': 'sum'}).reset_index()
trend_by_month = twitter_archive_master.groupby('year_month').agg({'tweet_id': 'count',
                                           'favorite_count': 'sum',
                                           'retweet_count': 'sum'}).reset_index()
```

```python
trend_by_date.plot(x='date', y='tweet_id')
trend_by_month.plot(x='year_month', y='tweet_id')
```

```python
trend_by_date.plot(x='date', y='favorite_count')
trend_by_month.plot(x='year_month', y='favorite_count')
```

```python
trend_by_date.plot(x='date', y='retweet_count')
trend_by_month.plot(x='year_month', y='retweet_count')
```

#### Q3: What are popular names people would get to their dog ?

The most common name people gave to their dogs are Copper, Charlie and Oliver, Then Tucker, lucy, Penny are the second tier. 

```python
base_color = sns.color_palette()[0]
plt.figure(figsize = (16, 12))
sns.countplot(data = twitter_archive_master, 
              y = twitter_archive_master.name,
              order = twitter_archive_master.name.value_counts()[:10].index, 
              color = base_color,
             );
for i in range(10):
    count = twitter_archive_master.name.value_counts()[i]
    pct_string = '{}'.format(count)
    plt.text(count + 0.1, i, pct_string, va='center')
    
plt.title('top 10 name')
```
